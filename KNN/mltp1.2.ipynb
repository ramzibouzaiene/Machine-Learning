```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
headernames = ['sepallength','sepalwidth','petallength','petalwidth','Class']
data = pd.read_csv("/content/iris.data", names = headernames)
array = data.values
X = array [ :, :2]
y = array [ :, 2]
data.shape
knnr = KNeighborsRegressor (n_neighbors = 10)
knnr.fit (X, y)
print ("Le MSE est : ", format (np.power (y-knnr.predict (X),2) .mean()))
```
